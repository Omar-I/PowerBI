# Power BI: introduction to DAX formulas

### Una vez que hayamos importado nuestra base de datos a Power BI, podremos utilizar "Power Query" y así poder editar las columnas de datos. 
### Es necesario eliminar la opción donde Power BI realiza en automatico dicha edición, ya que editaremos el formato de fecha y otras columnas.



### Para cambiar los tipos de datos, solo seleccionamos la casilla y nos desplegará los tipos de datos disponibles



### En la opción "Using Locale" se nos desplegará los formatos de datos disponibles en nuestra región, esto sirve, por ejemplo, para cambiar el tipo de formato de fecha de MM/DD/YY a DD/MM/YY




#### Al final se nos muestra el preview del cambio.



### En este apartado, se filtran datos que no vamos a requerir manipular (vamos a quitar a "United Kingdom"



#### Una vez hecho esto, sólo pulsamos en "Close and Apply" y los cambios serán guardados y a continuación podremos empezar a graficar nuestros datos.

#### En el apartado "Data" podemos hacer una manipulación de datos básica sin afectar nuestro modelo de datos, un análisis básico sería ordenar las "Invoice Date"(fecha de factura) para conocer la fecha donde comenzaron a facturarse.
